```python
from pprint import pprint

class Rider:
    def __init__(self, s):
        self.start_row, self.start_col, self.end_row, self.end_col, self.start, self.end = s
        self.length = abs(self.start_row-self.end_row) + abs(self.start_col - self.end_col)
    def __repr__(self):
        return '(%d,%d)->(%d,%d) [%d] at %d->%d [%d]' % (
            self.start_row, self.start_col, self.end_row, self.end_col,
            self.length,
            self.start, self.end, self.end - self.start
        )
    
    
f = open('./b_should_be_easy.in', 'r')
head = list(map(int, f.readline().split()))
print(head)
d = [Rider(list(map(int,l.split()))) for l in f]
#pprint(d)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18, 16))
for i, rider in enumerate(sorted(d, key=lambda r: r.start)):
    plt.plot([rider.start, rider.end], [i, i])
    
plt.show()
```
